```python
# Q.14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:
# (a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# (b) Label the x-axis as 'Date'.
# (c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import matplotlib.pyplot as plt

# Creating a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.randint(10, 30, size=30)
humidity = np.random.randint(40, 80, size=30)

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# Q.15  Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
#      tasks using Matplotlib:
# (a) Plot a histogram of the data with 30 bins.
# (b) Overlay a line plot representing the normal distribution's probability density function (PDF).
# (c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# (d) Set the title of the plot as 'Histogram

import matplotlib.pyplot as plt
from scipy.stats import norm

# Creating a NumPy array data containing 1000 samples from a normal distribution.
data = np.random.normal(loc=0, scale=1, size=1000)

# (a) Ploting a histogram of the data with 30 bins.
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# (b) Overlay a line plot representing the normal distribution's probability density function (PDF).
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

# (c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# (d) Set the title of the plot as 'Histogram with PDF Overlay'.
plt.title('Histogram with PDF Overlay')

# Displaying the plot.
plt.show()
```

```python
# Q.17  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import matplotlib.pyplot as plt
import seaborn as sns

# Creating two random arrays
x = np.random.rand(100) * 10 - 5  # Random numbers between -5 and 5
y = np.random.rand(100) * 10 - 5  # Random numbers between -5 and 5

# Determining quadrant for each point
quadrant = []
for i in range(len(x)):
   if x[i] >= 0 and y[i] >= 0:
       quadrant.append('Quadrant I')
   elif x[i] < 0 and y[i] >= 0:
       quadrant.append('Quadrant II')
   elif x[i] < 0 and y[i] < 0:
       quadrant.append('Quadrant III')
   else:
       quadrant.append('Quadrant IV')

# Creating the scatter plot
sns.scatterplot(x=x, y=y, hue=quadrant)

# Add legend, labels, and title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Displaying the plot
plt.show()
```

```python
# Q.18  Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

from bokeh.plotting import figure, show

import numpy as np

# Creating data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Creating a figure
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")

# Adding a line glyph
p.line(x, y, line_width=2)

# Adding grid lines
p.grid.grid_line_color = "gray"

# Showing the plot
show(p)
```

```python
# Q.20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

import plotly.express as px

# Generating random data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Creating the line plot
fig = px.line(x=x, y=y, title='Simple Line Plot')

# Updating axis labels
fig.update_layout(xaxis_title='X-axis', yaxis_title='Y-axis')

# Show the plot
fig.show()
```
